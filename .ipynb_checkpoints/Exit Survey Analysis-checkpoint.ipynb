# __DETE and TAFE Exit Survey Analysis__

In this project we will be analyzing exit survey from employees of the [Department of Education, Training and Employment]( https://qed.qld.gov.au/)(DETE) and the [Technical and Further Education]( https://tafeqld.edu.au/)(TAFE). 

The data for the DETE exit surveys can be found [here]( https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey), unfortunately the data for the TAFE exit survey data is no longer available online. 

In this project, we will attempt to answer the following questions: 
- Are employees who worked for the DOE for a short period of time resigning due dissatisfaction, why do long standing employees resign?
- Do employee age play as a factor in resigning due to dissatisfaction

We will begin the analysis by importing the libraires we will be using in this analysis: Pandas, NumPy and Matplotlib. Then we read our datasets using the pandas [`read_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

## __Creating the ` dissatisfied` column__

Now that we have an institute service for both of our data frames, we can analyze the survey respondents to see which employs resigned because they were dissatisfied.   

The columns that we will use to categorize employees as dissatisfied are:

`dete_resignation`:
- `job_dissatisfaction`
- `dissatisfaction_with_the_department`
- `physical_work_environment`
- `lack_of_recognition`
- `lack_of_job_security`
- `work_location`
- `employment_conditions`
- `work_life_balance`
- `workload`

`tafe_resignation`:
- `Contributing Factors. Dissatisfaction`
- `Contributing Factors. Job Dissatisfaction`

We will create a new column called `dissatisfied` which will contain `True` if the employee agreed with the question answered, `False` if they did not agree with any of the questions. We can see in the DETE survey the responses are Boolean, `True` and `False`. As we can see below in the pandas [`value_counts()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html) output bellow we can see TAFE survey however has `-` and column names as responses. 

```python
dete_col = ['job_dissatisfaction','dissatisfaction_with_the_department',
           'physical_work_environment', 'lack_of_recognition',
            'lack_of_job_security', 'work_location', 'employment_conditions',
           'work_life_balance', 'workload']

tafe_col = ['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']

print('tafe_resignation, Dissatisfaction')
print(tafe_resignation['Contributing Factors. Dissatisfaction'].value_counts())
print('\n')
print('tafe_resignation, Job Dissatisfaction')
print(tafe_resignation['Contributing Factors. Job Dissatisfaction'].value_counts())
```

To convert the values in the TAFE columns we will make a function that will convert `-` to `False`, positive responses to `True` and leave any `Null` values as `NaN`.  Once we have our function, we will apply it to the ` Contributing Factors. Dissatisfaction` and ` Contributing Factors. Job Dissatisfaction` using the pandas [`applymap()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.applymap.html?highlight=applymap#pandas.DataFrame.applymap) function.

```python
def update_vals(value):
    if pd.isnull(value):
        return np.nan
    elif value == '-':
        return False
    else:
        return True
     
tafe_resignation[tafe_col] = tafe_resignation[tafe_col].applymap(update_vals).astype(bool)
tafe_resignation[tafe_col].head()
```

```python
dete_resignation['dissatisfied'] = dete_resignation[dete_col].any(axis=1, skipna=False)

tafe_resignation['dissatisfied'] = tafe_resignation[tafe_col].any(axis=1, skipna=False)

dete_resignation_up = dete_resignation.copy()
tafe_resignation_up = tafe_resignation.copy()

display(dete_resignation_up['dissatisfied'].dtype)
display(tafe_resignation_up['dissatisfied'].dtype)
```

Now that we have all of the responses as Booleans, we can create our `dissatisfied` column. We can use the pandas [`any()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.any.html) function to see if there are any `True` values in the `dete_col` or the `tafe_col` and return the result into our new column. We will also copy our new data frames into new variabales usong the pandas [`copy()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.copy.html?highlight=copy#pandas.DataFrame.copy) function.

## Categorizing `age` and ` institute_service` columns 

### __Categorizing `age`__

The ` age` column now contains data from both of our dataframes, for us to continue our analysis we need to clean our values by convert them to numeric values to further manipulate the data. We can look at the unique ` age` values using the [`value_counts()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html?highlight=value_counts#pandas.Series.value_counts) function to see how to best clean it

```python
combined_drop['age'].value_counts()
```

```python
combined_drop['age'].isnull().sum()
```

We can see that in our ` age ` column we have a mix of values:

- numbers with a hyphen: `51-55, 26-30, 56-60`
- numbers with two spaces: `41  45, 21  26, 31  35`
- Sentences describing employee age: ’56 or older, 20 or younger’

we can convert all of these to floats by first converting the values to strings by the [`astype()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.astype.html?highlight=astype#pandas.Series.astype) and then cleaning the values with string manipulation method found [`here`](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html). We can the exclude our 55 NaN values as well by filtering for `!= ’nan’`

```python
combined_updated = combined_drop.copy()

combined_updated['age'] = (combined_updated['age'].astype(str)
                           .str.replace('56 or older', '56-56', regex=False)
                           .str.replace('61 or older', '61-61', regex=False)
                           .str.replace('20 or younger', '20-20', regex=False)
                           .str.replace('  ','-')
                           .str.strip())

combined_updated = combined_updated[combined_updated['age'] != 'nan']

combined_updated['age'].value_counts()
```

Now that we have our uniform ranges we can split the values using the vectorized [`split()`](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html) function to split the ranges on the `-`. Then we can save our two different numbers into two columns `lower_age` and `upper_age` which we can then convert into floats using the pandas [`astype()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html?highlight=astype#pandas.DataFrame.astype) function. 

```python
combined_updated['lower_age'] = (combined_updated['age']
                                 .str.split('-', expand = True)[0])
combined_updated['upper_age'] = (combined_updated['age']
                                 .str.split('-', expand = True)[1])

combined_updated['lower_age'] = combined_updated['lower_age'].astype(float)

combined_updated['upper_age'] = combined_updated['upper_age'].astype(float)

combined_updated.head()
```

We can now create a new column called ` age_average` where we can take the average of the ` lower_age` and ` upper_age` to get an age value for each employee. Once we have a value that we can analyze, we can drop the ` lower_age` and ` upper_age` columns using the pandas [`drop()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html?highlight=drop#pandas.DataFrame.drop) function to keep our dataframe clean. 

```python
combined_updated['age_average'] = (combined_updated['lower_age'] + combined_updated['upper_age']) /2 

combined_updated.drop(['lower_age', 'upper_age'], axis = 1, inplace = True)

combined_updated['age_average'].value_counts()
```

We can now make a function that will create a new column `age_cat` that will categorize an employee by their age in 10 year intervals. Once we have our function we can [`apply()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html?highlight=pandas%20apply#pandas.DataFrame.apply) to create our new column by applying the function to the `age` columns. 

```python
def age_cat(i):
    if i <30:
        return '20s'
    elif i < 40:
        return '30s'
    elif i < 50:
        return '40s'
    elif i < 60:
        return '50s'
    else:
        return '60s'
    
    
combined_updated['age_cat'] = combined_updated['age_average'].apply(age_cat)

combined_updated.head()
```

### __Categorizing `institute_service`__

The ` institute_service` column now contains data from both of our dataframes, for us to continue our analysis we need to clean our values by convert them to numeric values to further manipulate the data. We can look at the unique ` institute_service` values using the [`value_counts()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html?highlight=value_counts#pandas.Series.value_counts) function to see how to best clean it.

```python
combined_updated['institute_service'].value_counts()
```

We can see that in our ` institute_service` column we have a mix of values just like our `age` column:
- Numbers with decimal: `0.0, 5.0, 24.0`
- Number range with a hyphen: `1-2, 5-6, 11-20`
- Sentences describing employment length: `Less than 1 year, More than 20 years`

Like we did with the `age` we can convert all of these to floats by first converting the values to strings by the [`astype()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.astype.html?highlight=astype#pandas.Series.astype) and then cleaning the values with string manipulation method found [`here`](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html). We can the exclude any NaN values as well by filtering for `!= ’nan’`, then we finish by using the [`astype()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.astype.html?highlight=astype#pandas.Series.astype) to change all of the values back to a float. 


```python
combined_updated['institute_service'].value_counts()

combined_updadated_2 = combined_updated.copy()

combined_updadated_2['institute_service'] = (combined_updadated_2['institute_service'].astype(str)
                        .str.replace ('Less than 1 year', '1', regex = False)
                        .str.replace('More than 20 years', '20', regex = False)
                        .str.replace('1-2', '1.5', regex = False)
                        .str.replace('3-4', '3.5', regex = False)
                        .str.replace('5-6', '5.5', regex = False)
                        .str.replace('7-10', '8.5', regex = False)
                        .str.replace('11-20', '15.5', regex = False))

combined_updadated_2 = combined_updadated_2[combined_updadated_2['institute_service'] != 'nan']

combined_updadated_2['institute_service'] = combined_updadated_2['institute_service'].astype(float)

combined_updadated_2['institute_service'].unique()
```

Now that we have cleaned our ` institute_service` column we can create categories to describe employ type. In [`this`]( https://padillaco.com/wp-content/uploads/2018/06/padilla-engage-by-stage-research.pdf) article by  __SMS Research Advisors__, we can see that they have found that it is better to describe employees by career stage and address their needs by these categories. The SMS Research Advisors has defined the employee categories into 4 types:
- Newbie: 3 years or less at the company
- Sophomore: 4-7 years at the company
- Tenured: 8-10 years at the company
- Sage: 11 years or more at the company

We will create a simple function to assign a service category based on the ` institute_service` column. Once we have our function we can use the pandas [`apply()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html?highlight=series%20apply#pandas.Series.apply) function to create our new 'service_cat' column by applying the function to the ` institute_service` column.

```python
def i_range(i):
    if i < 3:
        return 'Newbie'
    elif i < 7:
        return 'Sophomore'
    elif i < 10:
        return 'Tenured'
    else:
        return 'Sage'
    
combined_updadated_2['service_cat'] = combined_updadated_2['institute_service'].apply(i_range)

combined_updadated_2.head()
```

Now that we have percentages of dissatisfied employees for each service category we can make a bar chart for our data using the [`matplotlib.pyplot.bar`]( https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.bar.html) function to visually inspect the values.
